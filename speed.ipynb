```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv("process_times.csv")

# Create a dictionary to map models and devices to colors and markers
model_colors = {"medium": "blue"}
device_markers = {
    "M2 Silicon": "o",
    "RTX 3060 Ti CUDA 12": "s",
}

# Plot the data
for (model, device), group in df.groupby(["Model", "Torch Device"]):
    plt.scatter(
        group["File Length (min)"],
        group["Process Time (min)"],
        color=model_colors[model],
        marker=device_markers[device],
        label=f"{model}, {device}",
    )

# Add labels and title
plt.xlabel("File Length (min)")
plt.ylabel("Process Time (min)")
plt.title("Process Time vs File Length")
plt.legend()

# Display the plot
plt.show()
```
